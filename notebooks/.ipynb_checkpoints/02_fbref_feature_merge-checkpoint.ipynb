## Merging all FBref Datasets

```python
shooting.columns.tolist()
```

```python
shooting = shooting.drop(
    columns=[
        "nineties",
        "goals",
        "penalty_goals",
        "penalty_attempts"
    ]
)
```

```python
merge_cols = [
    "league",
    "season",
    "team",
    "player",
    "nation",
    "position",
    "age",
    "born"
]

fbref = standard.merge(
    shooting,
    on=merge_cols,
    how="left"
)

fbref.shape
```

```python
fbref.columns
```

```python
playing_time.columns.tolist()
```

```python
playing_time = playing_time.drop(
    columns=[
        "matches_played",
        "starts",
        "minutes",
        "nineties"
    ]
)
```

```python
fbref = fbref.merge(
    playing_time,
    on=merge_cols,
    how="left"
)

fbref.shape
```

```python
fbref.columns.tolist()
```

```python
misc.columns = [
    "league",
    "season",
    "team",
    "player",
    "nation",
    "position",
    "age",
    "born",
    "nineties",
    "yellow_cards",
    "red_cards",
    "second_yellows",
    "fouls_committed",
    "fouls_drawn",
    "offsides",
    "crosses",
    "interceptions",
    "tackles_won",
    "penalty_wins",
    "penalty_conceded",
    "own_goals"
]
```

```python

misc = misc.drop(columns=[
    "nineties",
    "yellow_cards",
    "red_cards"
])

fbref = fbref.merge(
    misc,
    on=merge_cols,
    how="left"
)

fbref.shape
```

```python
fbref.shape
```

```python
list(enumerate(keeper.columns))
```

```python
keeper = keeper.drop(
    columns=[
        "matches_played",
        "starts",
        "minutes",
        "nineties"
    ]
)

fbref = fbref.merge(
    keeper,
    on=merge_cols,
    how="left"
)
```
