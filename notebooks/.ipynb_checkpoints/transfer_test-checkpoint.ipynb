## Custom Functions

```python
def clean_transfermarkt_table(df, club, season):
    
    df = df[df["Market value"].notna()].copy()
    
    df["club"] = club
    df["season"] = season
    
    return df
```

## Tackling Player Name Position Combination Error
- str.contains: https://www.w3schools.com/java/ref_string_contains.asp

```python
arsenal_check["Player"].tolist()


positions = [
    "Goalkeeper",
    "Centre-Back",
    "Left-Back",
    "Right-Back",
    "Defensive Midfield",
    "Central Midfield",
    "Attacking Midfield",
    "Right Midfield",
    "Left Midfield", # No left-mid
    "Left Winger",
    "Right Winger",
    "Centre-Forward"
]

for pos in positions:
    count = arsenal_check["Player"].str.contains(
        pos,
        regex=False
    ).sum()

    print(f"{pos}: {count}")
```

```python
# Extract Playing Positions from Player names

current_positions = [
    "Goalkeeper",
    "Centre-Back",
    "Left-Back",
    "Right-Back",
    "Defensive Midfield",
    "Central Midfield",
    "Attacking Midfield",
    "Right Midfield",
    "Left Winger",
    "Right Winger",
    "Centre-Forward"
]

def extract_position(text):
    """ A function that extracts a player's position from the Transfermarkt Player column.

        Parameters:
            text (str):
                A string containing both the player's name and position.

        Returns:
            str or None: The identified position if found. Returns None if no known position is detected.

    """
    
    for pos in current_positions:
        if text.endswith(pos):
            return pos
    return None



arsenal_check["position"] = (
    arsenal_check["Player"]
    .apply(extract_position)
)
```

```python
arsenal_check[
    ["Player", "position"]
].head(20)

#arsenal_check["Player"].head(20)
```

```python
def extract_player_name(text, position):
    """A function to extract the player's name from the Transfermarkt Player column.

        Parameters:
            text (str): Original Player column value containing both player name and position.

            position (str): Position extracted from the Player column.

        Returns:
            str: A clean player name with the position removed.
    """

    return text.replace(position, "").strip()


arsenal_check["player_name"] = arsenal_check.apply(
    lambda row: extract_player_name(
        row["Player"],
        row["position"]
    ),
    axis=1
)

arsenal_check[
    ["Player", "player_name", "position"]
].head(10)
```

## Convert Market value from characters to numbers
Reference: https://docs.adverity.com/reference/2-enrich/using-python-numerical.html

```python
def convert_market_value(value):
    """To convert Transfermarkt market values from strings into numeric values.

    Parameters:
        value (str): Player Market value string from Transfermarkt

    Returns:
        float: Market value in euros.
    """

    value = value.replace("€", "")

    if "m" in value:
        return float(value.replace("m", "")) * 1_000_000

    elif "k" in value:
        return float(value.replace("k", "")) * 1_000

    return None

arsenal_check["market_value_eur"] = (
    arsenal_check["Market value"]
    .apply(convert_market_value)
)

arsenal_check.head(10)
```

```python
arsenal_check[
    ["Market value", "market_value_eur"]
].head(6)
```

```python
arsenal_check[
    ["Player", "player_name", "position",
     "Market value", "market_value_eur"]
].head(15)
```

```python
club_links = soup.find_all("a")
len(club_links)
```

```python
for link in soup.find_all("a"):
    href = link.get("href")

    if href and "verein/" in href:
        print(href)
```

```python
import re

href = "/manchester-city/kader/verein/281/saison_id/2025"

match = re.search(r"verein/(\d+)", href)

print(match.group(1))
```

```python
club_links = []

for link in soup.find_all("a"):
    href = link.get("href")

    if href and "/kader/verein/" in href:
        club_links.append(href)

club_links[:10]
```

```python
unique_clubs = sorted(set(club_links))

len(unique_clubs)
```
