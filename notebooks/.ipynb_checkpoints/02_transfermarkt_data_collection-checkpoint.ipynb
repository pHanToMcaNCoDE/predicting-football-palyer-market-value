```python
from importlib import reload
import transfermarkt_utils

reload(transfermarkt_utils)
from transfermarkt_utils import *
```

```python
prem_url = (
    "https://www.transfermarkt.com/"
    "premier-league/startseite/wettbewerb/GB1/plus/?saison_id=2025"
)

prem_clubs = get_league_clubs(prem_url)
prem_clubs.head()
```

```python
len(prem_clubs)
```

```python
prem_df = collect_league_squads(
    prem_clubs,
    "2526"
)
```

```python
prem_df.shape
```

```python
prem_df.sample(20)
```

```python
prem_df["club"].nunique()
```

```python
url = prem_clubs.loc[
    prem_clubs["club_name"] == "Brentford Fc",
    "squad_url"
].iloc[0]

tables = pd.read_html(url)

tables[1].head(20)
```

```python
brentford_df = tables[1]

brentford_df[
    brentford_df["Market value"].notna()
][["Player", "Market value"]].tail(10)
```

```python
brentford_df["position"] = (
    brentford_df["Player"]
    .apply(extract_position)
)

brentford_df[
    brentford_df["position"].isna()
][["Player"]]
```

```python
prem_df.isna().sum()
```

```python
prem_df["position"].value_counts()
```

```python
prem_df.to_csv(
    "/Users/kachi/Documents/Thesis/data/raw/premier_league_2526.csv",
    index=False
)
```
